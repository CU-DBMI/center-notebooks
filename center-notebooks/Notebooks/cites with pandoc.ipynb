```python
%load_ext nb_black

import os, json
import pandoc
from manubot.cite.citations import Citations
```

```python
list_of_keys = ["pmc:PMC8525744", "pmc:PMC8108524", "pmc:PMC7780623"]
citations = Citations(list_of_keys)
cites = citations.get_csl_items()

cites
```

```python
current_path = os.getcwd()
doc = pandoc.read(
    source=json.dumps(cites),
    format="csljson",
    options=[
        "--citeproc",
        "--csl=manubot-style.csl",
        f"--resource-path={current_path}",
    ],
)
doc
```

```python
pandoc.write(
    doc,
    format="markdown",
    options=["--to=markdown_strict-raw_html", "--wrap=none", "--eol=lf"],
)
```
